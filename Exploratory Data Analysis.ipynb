# EXPLORATORY DATA ANALYSIS

What is exploratory Data analysis?
EDA is data exploration techniue to understand the various aspects of the data.

Objectie of EDA
It is basicall used to filter the data from redundancies
Helps us to understand the relationship between the variables

Steps Involved in EDA
Follows a systematic set of steps to explore the data in the most efficient way possible
1. Understand the Data
2. Clean the data from redundancies (irregularity in the data)
3. Analysis of relationship between variables
